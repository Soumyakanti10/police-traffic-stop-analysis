```python
import pandas as pd
```

```python
ri=pd.read_csv('police_dataset (1).csv')
ri
```

```python
#crosscheck the rows and columns
ri.shape
```

```python
ri.info()
```

```python
#create a backup to avoid operation in motherdata
ri_backup=ri
```

```python
ri_backup.head()
```

```python
#find the null value or empty space
ri_backup.isnull().sum()
```

```python
#drop the 'country name' and 'state'
ri_backup.drop(['county_name','state'],axis='columns',inplace=True)
```

```python
ri_backup
```

```python
#drop all rows that are missing gender
ri_backup.dropna(subset=['driver_gender'],inplace=True)
ri_backup
#dropna along with inplace=true going to comment empty row number and also help us to perform operation with skip of that particular row
```

```python
ri_backup.isnull().sum()
```

```python
#examine the head of the arrested people
ri_backup.is_arrested.head()
```

```python
#change the data type of 'is_arrested' to 'bool'
ri_backup['is_arrested'] = ri_backup.is_arrested.astype('bool')
ri_backup.info()
```

```python
#concatenate 'stop_date' and 'stop_time(separated by a space)
combined=ri_backup.stop_date.str.cat(ri_backup.stop_time,sep=' ')
combined
```

```python
#convert 'combined' to datetime format
ri_backup['stop_datetime']=pd.to_datetime(combined)
ri_backup
```

```python
#set 'stop_datetime' as the index ri_backup.set_index
ri_backup.set_index('stop_datetime',inplace=True)
ri_backup
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(x="violation", hue="driver_gender", data=ri)
plt.title("Violations by Gender")
plt.xticks(rotation=45)
plt.show()
```

```python
ri_backup.search_conducted.value_counts(normalize=True)
```

```python
ri_backup.search_conducted.mean()
```

```python
ri_backup.groupby('driver_gender').search_conducted.mean()
```

```python
#calculate the search rate for each combination of gender and violation
ri_backup.groupby(['driver_gender','violation']).search_conducted.mean()
```

```python
search_rate = ri.groupby("driver_gender")["search_conducted"].mean()

search_rate.plot(kind="bar", color=["#1f77b4","#ff7f0e"])
plt.title("Search Rate by Gender")
plt.ylabel("Search Rate")
plt.show()
```

```python
ri_backup['frisk']=ri_backup.search_type.str.contains('Protective Frisk',na=False)
ri_backup
```

```python
hourly_arrest_rate=ri_backup.groupby(ri_backup.index.hour).is_arrested.mean()
hourly_arrest_rate
```

```python
hourly_arrest_rate = ri.groupby(ri.index.hour)["is_arrested"].mean()

hourly_arrest_rate.plot(kind="line", marker="o")
plt.title("Hourly Arrest Rate")
plt.xlabel("Hour of Day")
plt.ylabel("Arrest Rate")
plt.grid(True)
plt.show()
```

```python
#calculate the monthly arrest rate
monthly_arrest_rate=ri_backup.groupby(ri_backup.index.month).is_arrested.mean()
monthly_arrest_rate

```
